```python
df=pd.get_dummies(df,drop_first=True)
df.columns
```

```python
df.diagnosis_M.dtype
```

```python
from scipy.stats import zscore
x=df.drop(labels='diagnosis_M',axis=1)
x=x.apply(zscore)
x.columns
```

```python
x=pd.DataFrame(x)
```

```python
df.diagnosis_M.dtype
```

```python
df['diagnosis_M']=df.diagnosis_M.dtype.astype('int64')
df.diagnosis_M.dtype
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
x=x.loc[:,['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
       'smoothness_mean', 'compactness_mean', 'concavity_mean', 'points_mean',
       'symmetry_mean', 'dimension_mean', 'radius_se', 'texture_se',
       'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se',
       'concavity_se', 'points_se', 'symmetry_se', 'dimension_se',
       'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst',
       'smoothness_worst', 'compactness_worst', 'concavity_worst',
       'points_worst', 'symmetry_worst', 'dimension_worst']]
y=df['diagnosis_M']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=1)

from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier 
import numpy as np
params={'n_neighbors':np.arange(1,30)}
knn=KNeighborsClassifier()
GS=GridSearchCV(knn,params,cv=5)
GS.fit(x,y)
k=GS.best_params_
k=k['n_neighbors']
print(' best n_neighbors value: ',k)



from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(k)
model.fit(x_train,y_train)


print('--------------------------------test_data------------------------------')
from sklearn import metrics
import numpy as np
y_prediction=model.predict(x_test)
fpr,tpr,t=metrics.roc_curve(y_test,y_prediction)
print(metrics.auc(fpr,tpr))
print('--------------------------------train_data------------------------------')
from sklearn import metrics
import numpy as np
y_prediction=model.predict(x_train)
fpr,tpr,t=metrics.roc_curve(y_train,y_prediction)
print(metrics.auc(fpr,tpr))



from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif["features"] = x.columns
vif.round(1)
```
